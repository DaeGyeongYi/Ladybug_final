```python
df = pd.read_csv("./일반화 과정//ursina_조사.csv")
```

```python
df = df.iloc[:,3:]
```

```python
df_max = df.iloc[:,8:]
```

```python
df_max
```

```python
df_max.describe()
```

```python
df_max['Year']= df['Year']
```

```python
for i in range(2007,2022):
    for j in df_max.columns:
        if df_max.loc[df_max['Year']==i,j].max()!=0 and j != 'Year':
            print(i,j,df_max.loc[df_max['Year']==i,j].max())
            df_max.loc[df_max['Year']==i,j] =  df_max.loc[df_max['Year']==i,j]/df_max.loc[df_max['Year']==i,j].max()
           
```

```python
df_max.describe()
```

```python
df_max['Source']=df['Source']
```

```python
df_max.to_csv("ursina조사_minmax,source.csv")
```
